```python
# Import section
import wordcloud
import nltk
from matplotlib import pyplot as plt
```

```python
def data_preprocessing(path):
    # Reading the data from the file:
    try:
        with open(path) as text:
            file_contents = text.read()
    except:
        print("Sorry, incorrect path or file format")

    # Generating dictionary with frequency of the words
    # Filtering "uninteresting" words and words with lengh is less than 3
    uninteresting_words = ["ours", "your", "yours", "hers", "they", "them", "their", "what", "which", "who", \
                        "whom", "this", "that", "were", "been", "being", "have", "does", "with", "from", \
                        "here", "when", "where", "both", "each", "more", "some", "such", "very", "will", "just", \
                        "there", "then", "would", "could", "page", "rowling", "didnt"]

    surname_list = ["Alice", "Harry", "Potter", "Dursley", "Hagrid", "Hermione", "Ron", "Snape",
                    "McGonagall", "Hogwarts", "Dumbledore", "Dudley", "Neville", "Vernon", "Weasley",
                     "Filch", "Malfoy", "Ron"]

    text_tokenized = nltk.word_tokenize(file_contents)
    text_tokenized = [word.lower() if word not in surname_list else word for word in text_tokenized]
    text_dict = nltk.probability.FreqDist(word for word in text_tokenized if len(word) > 3 and \
                                          word not in uninteresting_words or word == "Ron")
    
    # Preparing wordcloud from words dictionary
    cloud = wordcloud.WordCloud(width=800, height=400)
    cloud.generate_from_frequencies(text_dict)
    return cloud.to_array()
```

```python
def print_image(myimage):
    # Display your wordcloud image
    plt.figure( figsize=(20,10) )
    plt.imshow(myimage, interpolation = 'nearest')
    plt.axis('off')
    plt.show()
```

```python
# Initial data
path = r"./alice_in_wonderland.txt"

# Generating data for image:
myimage = data_preprocessing(path)

# Printing the result:
print_image(myimage)
```
